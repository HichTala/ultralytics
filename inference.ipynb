```python
import cv2
import numpy as np

from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO("./runs/detect/train14/weights/best.pt")

results = model(
    source="/home/cose-ia/Downloads/YuGiOh_YOLO.v2i.yolov8/video/2023-07-02 16-26-49 (online-video-cutter.com).mp4",
    show_labels=False,
    save=False
)
```

```python
colors = {
    'spell': np.array([120, 130, 40]),
    'monster': np.array([81, 86, 138]),
    'link': np.array([97, 66, 8]),
    'xyz': np.array([61, 46, 21]),
    'syncro': np.array([180, 173, 165]),
    'trap': np.array([124, 77, 133])
}

card_types = ['spell', 'monster', 'link', 'xyz', 'syncro', 'trap']

tol = {
    'spell': 35,
    'monster': 35,
    'link': 35,
    'xyz': 20,
    'syncro': 35,
    'trap': 35,
}
```

```python
spell_frame = 436
xyz_frame = 210
syncro_frame = 110
monster_frame = 523
link_frame = 612
trap_frame = 714
```

```python
result = results[trap_frame]
boxes = result[1]

img = result.orig_img.copy()
```

```python
x1, y1, x2, y2, = map(int, boxes.boxes.xyxy.squeeze())
roi = img[y1:y2, x1:x2].copy()

width = x2 - x1
height = y2 - y1

ones1 = np.ones(
    (int(0.71 * height) - int(0.18 * height), width - int(0.115 * width) - int(0.115 * width), 3)
)

ones2 = np.ones(
    (int(0.27 * height) - int(0.1 * height), width - int(0.115 * width) - int(0.115 * width), 3)
)

roi[
    height - int(0.71 * height): height - int(0.18 * height),
    int(0.115 * width):width - int(0.115 * width)
] = ones1 * [1000, 1000, 1000]

roi[
    int(0.1 * height):int(0.27 * height),
    int(0.115 * width):width - int(0.115 * width)
] = ones2 * [1000, 1000, 1000]
```

```python
roi = roi.reshape((roi.shape[0] * roi.shape[1], 3)).astype(float)

counts = [
    np.isclose(
        np.array([np.linalg.norm(diff) for diff in (roi - colors[card_type])]),
        np.zeros(roi.shape[0]),
        atol=tol[card_type]
    ).sum()
    for card_type in card_types
]
```

```python
counts
```

```python
spell = np.isclose(
    np.array([[np.linalg.norm(diff1) for diff1 in diff2] for diff2 in (roi - colors['spell'])]),
    np.zeros(roi.shape[0:2]),
    atol=35
)

monster = np.isclose(
    np.array([[np.linalg.norm(diff1) for diff1 in diff2] for diff2 in (roi - colors['monster'])]),
    np.zeros(roi.shape[0:2]),
    atol=35
)

link = np.isclose(
    np.array([[np.linalg.norm(diff1) for diff1 in diff2] for diff2 in (roi - colors['link'])]),
    np.zeros(roi.shape[0:2]),
    atol=35
)

xyz = np.isclose(
    np.array([[np.linalg.norm(diff1) for diff1 in diff2] for diff2 in (roi - colors['xyz'])]),
    np.zeros(roi.shape[0:2]),
    atol=20
)

syncro = np.isclose(
    np.array([[np.linalg.norm(diff1) for diff1 in diff2] for diff2 in (roi - colors['syncro'])]),
    np.zeros(roi.shape[0:2]),
    atol=35
)

trap = np.isclose(
    np.array([[np.linalg.norm(diff1) for diff1 in diff2] for diff2 in (roi - colors['trap'])]),
    np.zeros(roi.shape[0:2]),
    atol=35
)
```
